```python
import pandas as pd
import psycopg2 as psy
```

```python
df = pd.read_csv("./datasets/climber_df.csv")
```

```python
df.head()
```

```python
try:
    conn = psy.connect("host=127.0.0.1 dbname=imported_data_1 user=postgres password=root")
except pyscopg2.Error as e:
    print(e)
    
try:
    cursor = conn.cursor()
except psycopg2.Error as e:
    print(e)
    
conn.set_session(autocommit=True)
```

```python
for index, row in df.iterrows():
    insertQueryStr = "INSERT INTO climber_df (user_id, country, sex, height, weight, age, years_cl, grades_count, grades_first, grades_last, grades_max, grades_mean, year_first, year_last) VALUES ({}, '{}', '{}', {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {});"
    sex = "Male" if row["sex"] == 0 else "Female"
    insertQueryStr = insertQueryStr.format(row["user_id"], row["country"], sex, row["height"], row["weight"], row["age"], row["years_cl"], row["grades_count"], row["grades_first"], row["grades_last"], row["grades_max"], row["grades_mean"], row["year_first"], row["year_last"])
    cursor.execute(insertQueryStr)
#     print(insertQueryStr)
#     if index == 5:
#         break
```
